```python
import numpy as np
from matplotlib import pylab as plt
```

```python
p = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
T1 = 228.624
Tp = np.array([228.624, 114.345, 82.2178, 58.8667, 75.2502, 65.9119, 64.8308, 63.5097])
S = T1 / Tp
```

График

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S, color = 'orange')

plt.title("Ускорение от количества нитей", fontsize = 18)
plt.xlabel("Количество нитей", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.show()
```
