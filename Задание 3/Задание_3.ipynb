```python
import numpy as np
from matplotlib import pylab as plt
```

```python
p = [1, 2, 3, 4, 5, 6, 7, 8]
```

$N = 10^9$

```python
T1_1 = 143.309
Tp_1 = np.array([143.309, 71.6776, 48.3736, 37.5505, 39.7437, 39.0222, 36.5219, 37.3478])
S_1 = T1_1 / Tp_1
```

```python
T1_2 = 143.692
Tp_2 = np.array([143.692, 71.936, 56.753, 36.169, 42.5185, 36.1643, 37.6422, 36.446])
S_2 = T1_2 / Tp_2
```

```python
T1_3 = 38.1839
Tp_3 = np.array([38.1839, 18.5501, 18.8158, 12.7216, 9.56668, 8.27409, 7.64619, 7.45647])
S_3 = T1_3 / Tp_3
```

```python
T1_4 = 37.6575
Tp_4 = np.array([37.6575, 18.1913, 12.558, 9.26533, 8.72964, 7.72366, 6.97546, 6.73191])
S_4 = T1_4 / Tp_4
```

График

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S_1, color = 'orange', label = r'N = $10^9$ Кластер mutex')
plt.plot(p, S_3, color = 'red', label = r'N = $10^9$ Кластер pthread_exit – pthread_join')
plt.plot(p, S_2, color = 'green', label = r'N = $10^9$ Локально mutex')
plt.plot(p, S_4, color = 'blue', label = r'N = $10^9$ Локально pthread_exit – pthread_join')

plt.title("Ускорение от количества нитей", fontsize = 18)
plt.xlabel("Количество нитей", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.legend(fontsize = 15)
plt.show()
```
