```python
import numpy as np
from matplotlib import pylab as plt
```

```python
p = [1, 2, 3, 4, 5, 6, 7, 8]
```

__a)__ Для $N = 1000$

```python
T1_1 = np.array([4.79221e-05, 4.79221e-05, 5.31673e-05, 5.19753e-05, 5.4121e-05, 5.22137e-05, 5.50747e-05, 5.19753e-05])
Tp_1 = np.array([4.79221e-05, 6.19888e-05, 6.60419e-05, 0.000134945, 0.000329018, 0.000595093, 0.000478983, 0.00114083])
S_1 = T1_1 / Tp_1
```

__b)__ Для $N = 2 * 10^4$

```python
T1_2 = np.array([0.000963926, 0.000972986, 0.000988007, 0.00100112, 0.000996113, 0.000994205, 0.00107694, 0.001086])
Tp_2 = np.array([0.000963926, 0.000520945, 0.000448942, 0.000438929, 0.000566959, 0.00037694, 0.041676, 0.00125694])
S_2 = T1_2 / Tp_2
```

__b)__ Для $N = 10^6$

```python
T1_3 = np.array([0.0425138, 0.042773, 0.044107, 0.044184, 0.0442832, 0.0425792, 0.0439329, 0.0426159])
Tp_3 = np.array([0.0425138, 0.0213761, 0.0156941, 0.0118949, 0.00972295, 0.00767517, 0.007267, 0.00687599])
S_3 = T1_3 / Tp_3
```

__c)__ Для $N = 10^8$

```python
T1_4 = np.array([4.25875, 4.26475, 4.27304, 4.40582, 4.41896, 4.26264, 4.39504, 4.25753])
Tp_4 = np.array([4.25875, 2.13182, 1.42556, 1.09579, 0.933185, 0.714788, 0.669815, 0.555279])
S_4 = T1_4 / Tp_4
```

График

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S_1, color = 'orange', label = r'$N = 10^3$')
plt.plot(p, S_2, color = 'red', label = r'$N = 2 * 10^4$')
plt.plot(p, S_3, color = 'cyan', label = r'$N = 10^6$')
plt.plot(p, S_4, color = 'green', label = r'$N = 10^8$')

plt.title("Ускорение от количества процессов", fontsize = 18)
plt.xlabel("Количество процессов", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.legend(fontsize = 20)
plt.show()
```
