```python
import numpy as np
from matplotlib import pylab as plt
```

```python
p = [1, 2, 3, 4, 5, 6, 7, 8]
```

__a)__ Для $N = 2000$

```python
T1_11 = 0.0453529
Tp_11 = np.array([0.0453529, 0.0248761, 0.0194962, 0.0165629, 0.068574, 0.065551, 0.0719831, 0.070039])
S_11 = T1_11 / Tp_11
```

```python
T1_12 = 0.0452509
Tp_12 = np.array([0.0452509, 0.0248511, 0.0183249, 0.014962, 0.054368, 0.048311, 0.0522599, 0.0503721])
S_12 = T1_12 / Tp_12
```

__b)__ Для $N = 10000$

```python
T1_21 = 5.64088 
Tp_21 = np.array([5.64088, 2.88879, 1.99341, 1.59924, 2.5, 2.3548, 2.43893, 2.39399])
S_21 = T1_21 / Tp_21
```

```python
T1_22 = 5.66494
Tp_22 = np.array([5.66494, 2.8803, 2.00898,  1.59954, 2.49256, 2.33144, 2.83349,  2.32987])
S_22 = T1_22 / Tp_22
```

__b)__ Для $N = 50000$

```python
T1_31 = 7.07312
Tp_31 = np.array([7.07312, 3.55989, 2.40845, 1.85905, 1.84229, 1.53373, 1.5996, 1.32964])
S_31 = T1_31 / Tp_31
```

```python
T1_32 = 7.08669
Tp_32 = np.array([7.08669, 3.59073, 2.46598, 2.01883, 1.72075, 1.45745, 1.42133, 1.17658])
S_32 = T1_32 / Tp_32
```

График

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S_11, linestyle = '--', color = 'orange', label = 'Slow N = 2000')
plt.plot(p, S_12, color = 'orange', label = 'Fast N = 2000')

plt.plot(p, S_21, linestyle = '--', color = 'red', label = 'Slow N = 10000')
plt.plot(p, S_22, color = 'red', label = 'Fast N = 10000')

plt.plot(p, S_31, linestyle = '--', color = 'cyan', label = 'Slow N = 50000')
plt.plot(p, S_32,  color = 'cyan', label = 'Fast N = 50000')

plt.title("Ускорение от количества процессов", fontsize = 18)
plt.xlabel("Количество процессов", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.legend(fontsize = 20)
plt.show()
```
